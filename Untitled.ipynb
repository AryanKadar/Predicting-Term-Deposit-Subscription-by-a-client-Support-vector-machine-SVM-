```python
t=t.sample(6000)
```

```python
x=t.iloc[:,:14]
x
```

```python
y=t.iloc[:,14]
y
```

```python
xta,xte,yta,yte=TTS(x,y,test_size=0.2,random_state=3)
```

```python
ct=CT(
[
    ("OE",OE(categories=[job,default,month,d_w,poutcome]),[0,1,2,3,8]),
    
]
,remainder="passthrough"
)
```

```python
p={
    "C":[1,2,3,4],
    "kernel":["linear","poly","rbf","sigmoid"],
    "degree":[1,2,3],
    "gamma":["scale","auto"],
    "coef0":[0,1,2],
}
```

```python
pp1=PP(
[
    ("ct",ct),
    ("SS",SS()),
    ("svm",GSCV(svc(),param_grid=p,scoring="accuracy",n_jobs=-1,cv=3))
]
)
```

```python
pp1.fit(xta,yta)
```

```python
pp1.named_steps["svm"].best_score_
```
